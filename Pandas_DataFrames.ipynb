```python
import pandas as pd
import numpy as np
```

```python
data = {
    "name": ["Alice", "Bob", "Charlie", "David", "Eve","Alice"],
    "age": [30, 24, 45, np.nan, 29,30],
    "department": ["HR", "IT", "Finance", "IT", "HR","HR"],
    "salary": [60000, np.nan, 90000, 70000, 62000,60000]
}
df = pd.DataFrame(data)
df
```

```python
#inserting new date into a dataframe
df['promoted salary'] = df['salary'] * 10
df
```

```python
#Data cleaning

#->1.Checking null values in the dataframe on all columns.
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df
```

```python
df['Name']=df['Name'].replace('Charlie','Darsan')
```

```python
df
```

```python
#->2.Delaling with duplicate values...
df_dup = df[df.duplicated]
df_dup
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
# invalid values:
# Lambda -> python
df["promoted salary"] = df["promoted salary"].apply(lambda x: x/10 if x > 650000 else x)
```

```python
df
```

```python
#Joins and Merges
department_info ={
    "Dept":["HR","IT","Finance"],
    "Location":["New York","San Francisco", "Chicago"],
    "Manager":["Laura","Steve","Nina"]
}
```

```python
df2 = pd.DataFrame(department_info)
df2
```

```python
#Concat
pd.concat([df,df2],axis=1)
```

```python
#Merge
pd.merge(df,df2,on="Dept")
```
